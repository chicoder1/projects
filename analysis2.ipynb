```python
import pandas as pd
```

```python
df = pd.read_csv("/mnt/c/users/gautam/source/repos/VoltInsightsAutomation/Reports/data.csv")
```

```python
for col in df.select_dtypes([object]).columns[0:]:
    df[col] = df[col].fillna('0')
    df[col] = df[col].str.replace('$', '')
    df[col] = df[col].str.replace('%', '')
    df[col] = df[col].str.replace('(', '')
    df[col] = df[col].str.replace(')', '')
    df[col] = df[col].astype(float)
```

```python
df.dtypes
```

```python
def range(row):
    if row['Crew Margin Less Externals per Direct Manhour'] > 25:
        val = 'Greater than 25'
    elif row['Crew Margin Less Externals per Direct Manhour'] >= 10 and row['Crew Margin Less Externals per Direct Manhour'] <=25:
        val = '10 - 25'
    else:
        val = 'Less than 10'
    return val
    
```

```python
df['Crew Margin'] = df.apply(range, axis=1)
```
